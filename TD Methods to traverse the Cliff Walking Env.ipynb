### TD Prediction: State Values

In this section, we will write our own implementation of TD prediction (for estimating the state-value function).

We will begin by investigating a policy where the agent moves:
- `RIGHT` in states `0` through `10`, inclusive,  
- `DOWN` in states `11`, `23`, and `35`, and
- `UP` in states `12` through `22`, inclusive, states `24` through `34`, inclusive, and state `36`.

The policy is specified and printed below.  Note that states where the agent does not choose an action have been marked with `-1`.

````python
policy = np.hstack([1*np.ones(11), 2, 0, np.zeros(10), 2, 0, np.zeros(10), 2, 0, -1*np.ones(11)])
print("\nPolicy (UP = 0, RIGHT = 1, DOWN = 2, LEFT = 3, N/A = -1):")
print(policy.reshape(4,12))
````

Run the next cell to visualize the state-value function that corresponds to this policy.  Make sure that you take the time to understand why this is the corresponding value function!

````python
V_true = np.zeros((4,12))
for i in range(3):
    V_true[0:12][i] = -np.arange(3, 15)[::-1] - i
V_true[1][11] = -2
V_true[2][11] = -1
V_true[3][0] = -17

plot_values(V_true)
````

The above figure is what we will try to approximate through the TD prediction algorithm.

The algorithm for TD prediction has five arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `policy`: This is a 1D numpy array with `policy.shape` equal to the number of states (`env.nS`).  `policy[s]` returns the action that the agent chooses when in state `s`.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `V`: This is a dictionary where `V[s]` is the estimated value of state `s`.


````python
from collections import defaultdict, deque
import sys

def td_prediction(env, num_episodes, policy, alpha, gamma=1.0):
    # initialize empty dictionaries of floats
    V = defaultdict(float)
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        # Set initial state
        state = env.reset()
        while True:
            # Select the action based on the currert state and policy 
            action = policy[state]
            # Backup old value of V for the current state
            old_V = V[state]
            # Take the action to get next state and reward 
            next_state, reward, done, info = env.step(action)
            # Update the value for current state
            V[state] = ((1 - alpha) * old_V) + (alpha * (reward + (gamma * V[next_state])))
            state = next_state
            if done:
                break
    return V 
````

### TD Control: Sarsa

In this section, we will write our own implementation of the Sarsa control algorithm.

The algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.


````python
np.ones(env.nA) * 0.1/env.nA
````

````python
def get_epsilon_policy(env, Q_s, epsilon):
    policy = np.ones(env.nA) * epsilon / env.nA
    policy[np.argmax(Q_s)] = 1 - epsilon + (epsilon / env.nA) 
    return policy
````

````python
def sarsa(env, num_episodes, alpha, gamma=1.0):
    # initialize action-value function (empty dictionary of arrays)
    Q = defaultdict(lambda: np.zeros(env.nA))
    # initialize performance monitor
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()   
        
        # Set epsilon value for the current episode
        epsilon = 1/((i_episode)+1)
        
        # Get the initial state value
        state = env.reset()
        
        # Get the policy for the current state
        policy_s = get_epsilon_policy(env, Q[state], epsilon)
        
        # Choose action based on the current policy
        action = np.random.choice(np.arange(env.nA), p = policy_s)
        
        step = 0
        while True:
            # Take action to get next state
            next_state, reward, done, info = env.step(action)
            
            # If number of steps in the episode > 400, terminate episode (to stop from getting stuck in a loop)
            # In case of continuous tasks, no need for it
            if done or (step>400):
                step = 0
                break
            
            # Get policy for next state
            policy_s = get_epsilon_policy(env, Q[next_state], epsilon)
            
            # Get action based on the next state
            next_action = np.random.choice(np.arange(env.nA), p=policy_s)
            
            old_Q = Q[state][action]
            
            # Update Q
            Q[state][action] = ((1 - alpha) * old_Q) + (alpha * (reward + (gamma * Q[next_state][next_action])))
            
            state = next_state
            action = next_action
            step += 1
            
    return Q
````

````python
# obtain the estimated optimal policy and corresponding action-value function
Q_sarsa = sarsa(env, 5000, .01)

# print the estimated optimal policy
policy_sarsa = np.array([np.argmax(Q_sarsa[key]) if key in Q_sarsa else -1 for key in np.arange(48)]).reshape(4,12)
check_test.run_check('td_control_check', policy_sarsa)
print("\nEstimated Optimal Policy (UP = 0, RIGHT = 1, DOWN = 2, LEFT = 3, N/A = -1):")
print(policy_sarsa)

# plot the estimated optimal state-value function
V_sarsa = ([np.max(Q_sarsa[key]) if key in Q_sarsa else 0 for key in np.arange(48)])
plot_values(V_sarsa)
````

### Q-learning

In this section, we will write our own implementation of the Q-learning control algorithm.

The algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.


````python
def q_learning(env, num_episodes, alpha, gamma=1.0):
    # initialize empty dictionary of arrays
    Q = defaultdict(lambda: np.zeros(env.nA))
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        # Set Epsilon
        epsilon = 1/(i_episode + 1)
        
        # Get the first state of the episode
        state = env.reset()
        
        step = 0
        while True:
            policy_s = get_epsilon_policy(env, Q[state], epsilon)
            action = np.random.choice(np.arange(env.nA), p=policy_s)
            next_state, reward, done, info = env.step(action)
            if done or (step>300):
                step = 0
                break
            
            old_Q = Q[state][action]
            Q[state][action] = ((1 - alpha)*old_Q) + (alpha * (reward + np.max(Q[next_state])))
            step += 1
            state = next_state
    return Q
````

### TD Control: Expected Sarsa

In this section, we will write our own implementation of the Expected Sarsa control algorithm.

The algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.


````python
def expected_sarsa(env, num_episodes, alpha, gamma=1.0):
    # initialize empty dictionary of arrays
    Q = defaultdict(lambda: np.zeros(env.nA))
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        # Set Epsilon
        epsilon = 0.005
        
        # Get the first state of the episode
        state = env.reset()
        
        policy_s = get_epsilon_policy(env, Q[state], epsilon)
        
        step = 0
        while True:
            
            action = np.random.choice(np.arange(env.nA), p=policy_s)
            next_state, reward, done, info = env.step(action)
            if done or (step>300):
                step = 0
                break
                
            policy_s = get_epsilon_policy(env, Q[next_state], epsilon)
            old_Q = Q[state][action]
            Q[state][action] = ((1 - alpha)*old_Q) + (alpha * (reward + np.dot(policy_s , Q[next_state])))
            step += 1
            state = next_state
    return Q
````
